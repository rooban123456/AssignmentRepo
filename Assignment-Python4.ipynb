```python
class values:
    def __init__(self,sidea,sideb,sidec):
        self.sidea=sidea
        self.sideb=sideb
        self.sidec=sidec

class area(values):
    def __init__(self,*args):
        super().__init__(*args)
    def calcarea(self,sidea,sideb,sidec):
        sarea=int((self.sidea+self.sideb+self.sidec)/2)
        a=(sarea*(sarea-self.sidea)*(sarea-self.sideb)*(sarea-self.sidec)) ** 0.5
        return a
    def __str__(self):
        b=area.calcarea(self,self.sidea,self.sideb,self.sidec)
        return 'Area of this triangle is: ' + str(b)
    
    
```

```python
d=area(3,4,5)
print(d)
```

```python
def filter_long_words(text,num):
    a = text.split(',')
    return list(filter(lambda x: len(x) > num,a))
```

```python
filter_long_words('rooban,francisco,shiva,muthukrishnan',7)
```
