```python
def addit(a,b):
    return a + b
def subit(a,b):
    return a - b
def mulit(a,b):
    return a * b
def divit(a,b):
    return a / b
def figrt(a,b):
    if a > b:
        return a
    else:
        return b
def fisml(a,b):
    if a < b:
        return a
    else:
        return b

def myreduce(myfunction,series):
    c = myfunction(series[0],series[1])
    for i in range(2,len(series)):
        c = myfunction(c,series[i])
    return c

            
    
```

```python
def myfilter(calc,listitem):
    m=[]
    
    for i in range(len(calc)):
        
        if(calc[i]=='>'):
            opr='>'
            break
        elif(calc[i]=='<'):
            opr='<'
            break
        elif(calc[i]=='='):
            opr='='
            break
            
    a=list(calc.partition(opr))
    for j in range(len(listitem)):
        c=a[0]
        if (len(a[0])>1):
            d=c.replace('x',str(listitem[j]))
            e=d.partition('+')
            g=d.partition('*')
            h=d.partition('-')
            if (e[1] == '+'):

                f=int(d[0]) + int(d[2])
                if(opr == '='):
                    if (f==int(a[2])):
                        m.append(listitem[j])
                if(opr == '>'):
                    if (f>int(a[2])):
                        m.append(listitem[j])
                if(opr == '<'):
                    if (f<int(a[2])):
                        m.append(listitem[j])

            if (h[1] == '-'):

                f=int(d[0]) - int(d[2])
                if(opr == '='):
                    if (f==int(a[2])):
                        m.append(listitem[j])
                if(opr == '>'):
                    if (f>int(a[2])):
                        m.append(listitem[j])
                if(opr == '<'):
                    if (f<int(a[2])):
                        m.append(listitem[j])

            if (g[1] == '*'):

                f=int(d[0]) * int(d[2])
                if(opr == '='):
                    if (f==int(a[2])):
                        m.append(listitem[j])
                if(opr == '>'):
                    if (f>int(a[2])):
                        m.append(listitem[j])
                if(opr == '<'):
                    if (f<int(a[2])):
                        m.append(listitem[j])
            
        else:
            d=c.replace('x',str(listitem[j]))
            f=int(d)
            if(opr == '='):
                if (f==int(a[2])):
                    m.append(listitem[j])
            if(opr == '>'):
                if (f>int(a[2])):
                    m.append(listitem[j])
            if(opr == '<'):
                if (f<int(a[2])):
                    m.append(listitem[j])
            
    return m
            
```
